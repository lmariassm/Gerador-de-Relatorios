CURVA DE TENDÊNCIA UTILIZANDO AS FUNÇÕES C(t)

```python
import math as mt
import matplotlib.pyplot as plt
#entrada de dados #CSTR
a=1 #ordem da reação
#F0=F1=F2=F
F = 5 #l/min
V = 10 #l
Ca0 = 1.8
K= 0.5 #k1=k2=k3=K
tau = 2
h=0.1 #passo
ti = 0
tf= 20 #chute

#FUNÇÕES
#tanque 1
def CA1 (t):
    return 0.5-(mt.exp(-t))/2
#tanque 2
def CA2 (t):
    return 0.25-(mt.exp(-t))/4-(t*mt.exp(-t))/4
#tanque 3
def CA3 (t):
    return 0.125-(mt.exp(-t))/8-(t*mt.exp(-t))/8-(t**2*mt.exp(-t))/16

#Armazenando valores em uma lista
x_t=[]
y_Ca1=[]
y_Ca2=[]
y_Ca3=[]

for i in range (0,int(tf/h)):
    x_t.append(ti)
    y_Ca1.append(CA1(ti))
    y_Ca2.append(CA2(ti))
    y_Ca3.append(CA3(ti))
    ti= ti + h

#Plot gráfico
fig, ax = plt.subplots()
plt.plot(x_t, y_Ca1, c = (.10, 0.5, .21), lw = 1., label = 'CA1')
plt.plot(x_t, y_Ca2, c = (.60, 1.0, .15), lw = 1., label = 'CA2')
plt.plot(x_t, y_Ca3, c = (.80, 0.0, .21), lw = 1., label = 'CA3')
ax.set(xlabel='Tempo', ylabel='Concentração')
ax.grid()
plt.legend()
plt.axis([0,20,0,0.6])
plt.title('Parâmetros cinéticos - Degrau unitário')
plt.show()

```

MAIS E MENOS 10% DA PERTURBAÇÃO DEGRAU

```python
from inspect import EndOfBlock
import matplotlib.pyplot as plt

#entrada de dados #CSTR
a=1 #ordem da reação
#F0=F1=F2=F
F = 5 #l/min
V = 10 #l
Ca0 = 1.8
K= 0.5 #k1=k2=k3=K
tau = 2
h=0.1 #passo
ti = 0
tf= 20 #chute

#Estimativa 1
Ca1 = 0.4
Ca2 = 0.2
Ca3 = 0.1
Ca1_ = 0.4
Ca2_ = 0.2
Ca3_ = 0.1

#FUNÇÕES DE TRANSFERÊNCIA
def dCA1(Ca0,Ca1):
    return (1/tau)*(Ca0 - Ca1) - K*Ca1

def dCA2(Ca1,Ca2):
    return (1/tau)*(Ca1 - Ca2) - K*Ca2

def dCA3(Ca2,Ca3):
    return (1/tau)*(Ca2 - Ca3) - K*Ca3

#Armazenando valores em uma lista
x_t=[]
#+10%
y_Ca1=[]
y_Ca2=[]
y_Ca3=[]
#-10%
y__Ca1=[]
y__Ca2=[]
y__Ca3=[]

for i in range (0,int(tf/h)):
    x_t.append(h*i)
    if h*i <= 7.5:
        Ca0_ = Ca0
        Ca0__ = Ca0
    else:
        Ca0_ = Ca0*1.1
        Ca0__ = Ca0*0.9

    #+10%
    Ca1 = Ca1 + h*dCA1(Ca0_, Ca1)
    Ca2 = Ca2 + h*dCA2(Ca1, Ca2)
    Ca3 = Ca3 + h*dCA3(Ca2, Ca3)
    #-10%
    Ca1_ = Ca1_ + h*dCA1(Ca0__, Ca1_)
    Ca2_ = Ca2_ + h*dCA2(Ca1_, Ca2_)
    Ca3_ = Ca3_ + h*dCA3(Ca2_, Ca3_)
    y_Ca1.append(Ca1)
    y_Ca2.append(Ca2)
    y_Ca3.append(Ca3)
    y__Ca1.append(Ca1_)
    y__Ca2.append(Ca2_)
    y__Ca3.append(Ca3_)

#Plot gráfico
fig, ax = plt.subplots()
plt.plot(x_t, y_Ca1, lw = 1., label = 'Ca1 +10%')
plt.plot(x_t, y_Ca2, lw = 1., label = 'Ca2 +10%')
plt.plot(x_t, y_Ca3, lw = 1., label = 'Ca3 +10%')
plt.plot(x_t, y__Ca1, lw = 1., label = 'Ca1 -10%')
plt.plot(x_t, y__Ca2, lw = 1., label = 'Ca2 -10%')
plt.plot(x_t, y__Ca3, lw = 1., label = 'Ca3 -10%')
ax.set(xlabel='Tempo', ylabel='Concentração')
plt.axis([0,20,0,1])
ax.grid()
plt.legend()
plt.title('Comparação das respostas às perturbações degrau')
plt.show()
```
